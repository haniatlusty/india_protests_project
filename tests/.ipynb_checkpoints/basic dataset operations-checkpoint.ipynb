```julia
using StatsKit, Plots; gr();
```

```julia
cities = CSV.File("../data/2016-2022.csv",select=["location"]).location;
```

```julia
ct        = countmap(cities);
ct_sorted = sort(collect(ct), by=x->x[2], rev=true);
```

```julia
data   = collect.(ct_sorted[1:4])
```

```julia
l      = length(cities);
names  = first.(data);
n      = last.(data)./l;
push!(names,"Other");
push!(n, 1-sum(n));
```

```julia
pie(names,n, 
    legend=:topleft,
    size=(1200,600),
    palette=cgrad(:lightrainbow, categorical=true,rev=true),
    linecolor = :white,
    linewidth = 0
)
```
